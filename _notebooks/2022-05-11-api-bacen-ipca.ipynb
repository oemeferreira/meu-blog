```python
import requests
import datetime
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# URL dos dados da API
url = 'http://api.bcb.gov.br/dados/serie/bcdata.sgs.16121/dados?formato=json'
```

```python
# Lendo a URL e coletando em um Dataframe
df = pd.read_json(url)
print(df)
```

```python
# Salva o DataFrame em um arquivo csv
df.to_csv("dados_ipca.csv", index=False)
```

```python
#Gráfico dos últimos 12 meses
meses = df['data'][-12:]
valores = df['valor'][-12:]

plt.plot(meses, valores, 'r-o')
plt.xticks(rotation=45)
plt.show()
```
